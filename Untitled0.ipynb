```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores.faiss import FAISS
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
```

### 外部データ読込み

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("./contents/hashire_merosu.txt")
doc = loader.load()
```

```python
print(doc)
```

### チャンクに分割

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
splits = text_splitter.split_documents(doc)
```

```python
splits[0:3]
```

### ベクターストアの作成・保存

```python
vectorstore = FAISS.from_documents(splits, OpenAIEmbeddings())
vectorstore.save_local("./vectorstore")
```

```python
vectorstore = FAISS.load_local("./vectorstore", embeddings=OpenAIEmbeddings())
```

```python
vectorstore.similarity_search("メロスを殴ったのは誰？")
# vectorstore.similarity_search_with_score("メロスを殴ったのは誰？")
```

```python
from langchain.chains import RetrievalQA
from langchain.prompts.prompt import PromptTemplate
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0) # "gpt-4", "gpt-3.5-turbo",
retriever = vectorstore.as_retriever()

qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type='stuff',
    retriever=retriever,
)
```

```python
qa.invoke("メロスを殴ったのは誰？")
```

```python
qa.invoke("セリヌンティウスを殴ったのは誰？")
```
